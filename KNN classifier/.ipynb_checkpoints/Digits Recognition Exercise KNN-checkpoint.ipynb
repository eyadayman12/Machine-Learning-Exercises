### Exercise
From sklearn.datasets load digits dataset and do following <br>
 1. Classify digits (0 to 9) using KNN classifier. You can use different values for k neighbors and need to figure out a value of K that gives you a maximum score. You can manually try different values of K or use gridsearchcv<br>
 2. Plot confusion matrix<br>
 3. Plot classification report

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from seaborn import heatmap
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import joblib
```

#### Exploaring the data

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df.head()
```

```python
df['Target'] = digits.target
df.head()
```

#### assign feature and target variable

```python
X = df.drop('Target', axis=1)
y = df.Target
```

#### Spliting the data into training and testing data

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)
```

#### Bulding and training the model

```python
model_params = {
    'KNN' : {
        'model' : KNeighborsClassifier(),
        'params' : {
            'n_neighbors' : np.arange(5,100,5),
            'leaf_size' : np.arange(10,100,10),
            'n_jobs' : ['None', -1]
        }
    }
}
```

```python
scores = []
for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(X_train, y_train)
    scores.append({
        'model' : model_name,
        'best_score' : clf.best_score_,
        'best_params' : clf.best_params_
    })
```

```python
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

best params leaf_size = 10 , n_jobs = -1, n_neighbors = 5

#### Accuracy of the model

```python
clf.score(X_test, y_test)
```

#### Confuesion Matrix

```python
y_predicted = clf.predict(X_test)
cm = confusion_matrix(y_predicted, y_test)
cm
```

```python
plt.figure(figsize=(10,8))
heatmap(cm, annot=True)
plt.show()
```

#### Classification report

```python
class_rep = classification_report(y_test, y_predicted)
print(class_rep)
```

#### save the model

```python
joblib.dump(clf, 'Digits Recogniation')
```

#### load the model

```python
Digits_Recogniation = joblib.load('Digits Recogniation')
```

#### Using the model after loading it

```python
Digits_Recogniation.predict(X_test)
```
